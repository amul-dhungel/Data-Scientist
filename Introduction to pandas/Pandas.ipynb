# Dictionary to DataFrame
In pandas there is a way of representing multiple dictionaries into dataframe.

```python
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
```

```python
# Import pandas as pd
import pandas as pd
```

```python
# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'country': names, 'drives_right': dr, 'cars_per_cap': cpc}
my_dict
```

```python
# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'country': names, 'drives_right': dr, 'cars_per_cap': cpc}

# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict)

cars
```

## Indexing of table with certain data

```python
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']
# Specify row labels of cars
cars.index = row_labels

# Print cars again
cars
```

```python
## Reading of CSV file in pandas
url = 'https://raw.githubusercontent.com/AmoDinho/datacamp-python-data-science-track/master/Intermediate%20Python%20for%20Data%20Science/Datasets/cars.csv'

# creating of first column as an index of a table
cars = pd.read_csv(url,index_col=0)
cars
```

## Subsetting of dataframe with the use of loc,iloc and square brackets
Subsetting means the dissecting of whole tables into different form, like
outputting only columns or specific rows

```python
# single brack gives series
cars["country"]
```

```python
# double bracket gives dataframe
cars[["country"]]
```

```python
# print out first three observations of rows
cars[0:3]
```

```python
# loc is to select data in the form of specififying row label
cars.loc[['AUS','JAP']]
```

```python
# subsetting of specific rows with desire columns
# first bracket is for rows, second is for columns
cars.loc[['RU', 'MOR'], ['country', 'drives_right']]
```

```python
# Print out drives_right column as DataFrame
print(cars.iloc[:, [2]])

# to print out all rows but specific columns only
cars.loc[:, ['cars_per_cap', 'drives_right']]
```

## Filtering dataframe
This is a technique of filter dataframe by using logical operations like outputting only values of columns with "True"

```python
# Extract drives_right column
dr = cars['drives_right']

# Use dr to subset cars with only "True values"
sel = cars[dr]

# Print sel,which outputs the only columns wiht true values
sel
```

```python
# converting of above code into one linear form
dr = cars[cars['drives_right']]
dr
```

```python
# Create car_maniac: observations that have a cars_per_cap over 500
car_maniac = cars[cars['cars_per_cap'] > 500]

# Print car_maniac
car_maniac
```

```python
import numpy as np

# Create medium: observations with cars_per_cap between 100 and 500
cpc = cars['cars_per_cap']
# to evaluate a single column with more than one evaluation , we should used numpy function
between = np.logical_and(cpc > 100, cpc < 500)
medium = cars[between]

# Print medium
medium
```

## Slicing and indexing dataframes

```python
# Indexing of table is necessary, because it makes easy in subsetting 

temperatures = pd.read_csv('https://raw.githubusercontent.com/amul-dhungel/Data-Scientist/main/Introduction%20to%20pandas/temperatures.csv')
temperatures.head(10)
```

```python
# Indexing of table by sepecific columns
# Index temperatures by city
temperatures_ind = temperatures.set_index("city")

# Look at temperatures_ind
temperatures_ind
```

```python
# resetting of index to arrange in proper way
temperatures_ind.reset_index()
```

```python
# dropping of index is like dropping of column
(temperatures_ind.reset_index(drop=True))
```

```python
# Index table by more than one columns
temperatures_ind = temperatures.set_index(["country", "city"])

# List of tuples: Brazil, Rio De Janeiro & Pakistan, Lahore
rows_to_keep = [("Brazil", "Rio De Janeiro"), ("Pakistan", "Lahore")]

# Subset for rows to keep, is showing of rows containing specific row data of two indices
temperatures_ind.loc[rows_to_keep]
```

```python
# to get range of data in row, first the table need to be sorted based on index columns.
# only sorted columns data will follow the .loc data range format.

# Index temperatures by country & city
temperatures_srt = temperatures_ind.sort_index()


# Subset for specific range of rows
temperatures_srt.loc["Pakistan":"Russia"]
```

```python
# specific
temperatures_srt.loc[("Pakistan", "Lahore"):("Russia", "Moscow")]
```

```python
temperatures_srt.head()
```

```python
# Subset rows from India, Hyderabad to Iraq, Baghdad
print(temperatures_srt.loc[("India", "Hyderabad"):("Iraq", "Baghdad")])

# Subset columns from date to avg_temp_c
print(temperatures_srt.loc[:, "date":"avg_temp_c"])

# Subset in both directions at once
print(temperatures_srt.loc[("India", "Hyderabad"):("Iraq", "Baghdad"), "date":"avg_temp_c"])
```

```python
 #sorting is necessary when you want to take out range of values from ascending to descending through index
temperature_time = temperatures.set_index('date').sort_index()
temperature_time
```
